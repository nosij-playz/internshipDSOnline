```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

Load Dataset

```python
df=pd.read_csv("diabetes.csv")
```

```python
df.head()
```

Prepare Data (Clean from Day 4)

```python
cols=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for c in cols:
    df[c]=df[c].replace(0,np.nan)
    df[c]=df[c].fillna(df[c].median())
```

```python
df=df.drop_duplicates()
```

```python
df.head()
```

Task 1: Separate Features and Target

```python
X=df.drop('Outcome',axis=1)
```

```python
X.head()
```

```python
y=df['Outcome']
```

```python
y.head()
```

Task 2: Check data types for encoding

```python
X.dtypes
```

```python
X.info()
```

Task 3: Apply Feature Scaling (StandardScaler)

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X.columns)
```

```python
X_scaled.head()
```

```python
X_scaled.describe()
```

Task 4: Split data into training and testing sets

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

Task 5: Train Logistic Regression Model

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

Task 6: Make Predictions

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_train_pred=model.predict(X_train)
```

```python
y_train_pred
```

Task 7: Evaluate Model Performance

```python
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_train,y_train_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

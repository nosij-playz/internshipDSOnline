Task 2: Identify Skewness and Outliers

### Skewness Analysis

```python
df.skew(numeric_only=True)
```

```python
df['Performance Index'].skew()
```

```python
df['Hours Studied'].skew()
```

```python
df['Previous Scores'].skew()
```

```python
print("Skewness Analysis:")
print("="*50)
score_cols=['Performance Index','Hours Studied','Previous Scores']
for col in score_cols:
    skewness=df[col].skew()
    print(f"{col}: {skewness:.4f}")
    if abs(skewness)<0.5:
        print(f"  -> Fairly symmetrical")
    elif abs(skewness)<1:
        print(f"  -> Moderately skewed")
    else:
        print(f"  -> Highly skewed")
```

Detect outliers using IQR method

```python
Q1=df['Performance Index'].quantile(0.25)
Q3=df['Performance Index'].quantile(0.75)
IQR=Q3-Q1
```

```python
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
```

```python
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

```python
outliers_perf=(df['Performance Index']<lower_bound)|(df['Performance Index']>upper_bound)
```

```python
outliers_perf.sum()
```

```python
for col in score_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers=(df[col]<lower_bound)|(df[col]>upper_bound)
    print(f"{col}: {outliers.sum()} outliers")
```

Prepare Features and Target

```python
X=df.drop('Performance Index',axis=1)
```

```python
X.head()
```

```python
y=df['Performance Index']
```

```python
y.head()
```

Task 5: Apply Feature Encoding

```python
X.dtypes
```

```python
categorical_features=X.select_dtypes(include='object').columns.tolist()
numerical_features=X.select_dtypes(include=['int64','float64']).columns.tolist()
```

```python
categorical_features
```

```python
numerical_features
```

### Option 1: One-Hot Encoding using pd.get_dummies

```python
X_encoded=pd.get_dummies(X,columns=categorical_features,drop_first=True)
```

```python
X_encoded.head()
```

```python
X_encoded.shape
```

```python
X_encoded.columns
```

### Option 2: Label Encoding (for demonstration)

```python
X_label=X.copy()
```

```python
le=LabelEncoder()
```

```python
for col in categorical_features:
    X_label[col]=le.fit_transform(X_label[col])
```

```python
X_label.head()
```

```python
X_label.dtypes
```

Using One-Hot Encoded features for further processing

```python
X=X_encoded
```

```python
X.head()
```

Task 4: Apply Feature Scaling

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X.columns)
```

```python
X_scaled.head()
```

```python
X_scaled.describe()
```

```python
print("Before Scaling:")
print(X[numerical_features].describe())
print("\nAfter Scaling:")
print(X_scaled[numerical_features].describe())
```

Task 6: Split Data into Training and Testing Sets

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
print("="*60)
print("DATA PREPARATION SUMMARY")
print("="*60)
print(f"Total samples: {len(df)}")
print(f"Training samples: {X_train.shape[0]} ({X_train.shape[0]/len(df)*100:.1f}%)")
print(f"Testing samples: {X_test.shape[0]} ({X_test.shape[0]/len(df)*100:.1f}%)")
print(f"\nFeatures after encoding: {X_scaled.shape[1]}")
print(f"Original categorical features: {len(categorical_features)}")
print(f"Numerical features: {len(numerical_features)}")
print(f"\nTarget variable: Performance Index")
print(f"Target range: {y.min()} to {y.max()}")
print("="*60)
print("Data is ready for model training!")
print("="*60)
```

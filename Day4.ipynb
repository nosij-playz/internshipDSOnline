Import Libraries

```python
import pandas as pd
```

```python
import numpy as np
```

Load Dataset

```python
df=pd.read_csv("diabetes.csv")
```

```python
df.head()
```

Task 1: Replace invalid zeros with NaN

```python
cols=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for c in cols:
    df[c]=df[c].replace(0,np.nan)
```

```python
df[cols].isnull().sum()
```

Task 2: Handle missing values (median)

```python
for c in cols:
    df[c]=df[c].fillna(df[c].median())
```

```python
df[cols].isnull().sum()
```

Task 3: Verify cleaned data

```python
df.describe()
```

```python
(df[cols]==0).sum()
```

```python
df.isnull().sum()
```
